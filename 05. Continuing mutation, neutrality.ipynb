# The organism class
First, let's consider an organism with a 'genetic code' of bits

```python
from scipy.stats import bernoulli
class organism:
    """An example of a simple organism"""
    def __init__(self,given_genome=[]):
        """Initialize a genome with a random collection of bits"""
        if len(given_genome) == 0:
            self.genome = bernoulli.rvs(.5,size=30)
        else:
            self.genome = given_genome
    def replicate(self, u = 0):
        """make a copy of a genome, possibly with error"""
        errors = bernoulli.rvs(u,size=len(self.genome))
        new_genome = []
        for i in range(len(self.genome)):
            if errors[i]:
                new_genome.append(int(not self.genome[i]))
            else:
                new_genome.append(int(self.genome[i]))
        return organism(new_genome)
myOrg = organism()
print list(myOrg.genome)
print myOrg.replicate(u=.1).genome
```

```python
import matplotlib.pyplot as plt
import numpy as np
myOrg = organism(np.array([1]*30))
critters = np.array(myOrg.genome)
for i in range(10):
    critters = np.vstack([critters, myOrg.replicate(.1).genome ])

%pylab inline
fig, ax = plt.subplots()
ax.set_xlabel('Parental genotype is at the bottom', fontsize=20)
plt.ylim((0,11))
heatmap = ax.pcolor(critters, cmap=plt.cm.Blues)
```

In the plot above, the original genotype is in blue, while the mutants are in white

# Thought experiment
  - Let's imagine we have a very long genome
    - So long, that the probability of the same mutation occuring twice at the same location is close to zero
       - by the way, a given location in the genome is called a **locus**
  - The population of organisms is finite
  - Oranisms are sampled with replacement from one generation to the next
  - Mutations are introduced with rate *u*

What is the likelihood of a mutation becoming *fixed* in the population (*i.e.*, to replace all of the ancestral versions).

  - by the way, different versions of the same gene are known as **alleles**
  - what corresponds to alleles in our toy example?

### Simulating the thought experiment

```python
from random import randint
from __future__ import division
def drift(popsize = 10,numgen = 100,startM = 1):  
    # inspired by from http://blog.drewsday.com/2012/08/a-simple-genetic-drift-simulation.html
    """ takes a fixe population size, the number of generations, and a number of mutants.
    Simulates a population sampled with replacement \
    Returns mutant frequencies over the generations"""
    freqM = [startM/popsize]
    prevgen = [1]*startM + [0]*(popsize - startM)
    for gen in range(numgen):
        nextgen = []
        for ind in range(popsize):
            nextgen = nextgen + [prevgen[(randint(0,popsize-1))]]
        freqM.append(sum(nextgen)/popsize)
        prevgen = nextgen
    return(freqM)
```

```python
#import matplotlib.pyplot as plt
for i in range(10):
    plt.plot(drift(numgen=50,popsize=10))
plt.show()
```
